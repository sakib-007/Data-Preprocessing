```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer

```

```python
data = pd.read_csv("/content/Income Dataset (50k).csv")
data
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
impute = SimpleImputer(missing_values=np.nan, strategy='mean')

impute.fit(data[['fnlwgt']])
impute.fit(data[['educational-num']])

data['fnlwgt'] = impute.transform(data[['fnlwgt']])
data['educational-num'] = impute.transform(data[['educational-num']])
data
```

```python
data.drop(['race'], axis=1, inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
data["gender"] = enc.fit_transform(data["gender"])
data[["gender"]].head()
data

#data["gender"] = data["gender"].map({"Male":1 , "Female":0})
#data
```

```python
from sklearn.model_selection import train_test_split
x = data[["age" , "fnlwgt" , "educational-num" , "hours-per-week" , "capital-gain" , "capital-loss"]]
y = data[["income_>50K"]]
X_train, X_test, y_train, y_test = train_test_split(x , y , test_size = 0.25, random_state = 10)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
print("per-feature minimum before scaling:\n {}".format(X_train.min(axis=0)))
print("per-feature maximum before scaling:\n {}".format(X_train.max(axis=0)))
print("per-feature minimum after scaling:\n {}".format(
    X_train_scaled.min(axis=0)))
print("per-feature maximum after scaling:\n {}".format(
    X_train_scaled.max(axis=0)))
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
x = data[["age" , "fnlwgt" , "educational-num" , "hours-per-week" , "capital-gain" , "capital-loss"]]
y = data[["income_>50K"]]
X_train, X_test, y_train, y_test = train_test_split(x , y , test_size = 0.25, random_state = 10)
knn=KNeighborsClassifier()

knn.fit(X_train, y_train)

print("Test set accuracy: {:.2f}".format(knn.score(X_test, y_test)))
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)

X_test_scaled = scaler.transform(X_test)

knn.fit(X_train_scaled, y_train)
print("Scaled test set accuracy: {:.2f}".format(
    knn.score(X_test_scaled, y_test)))
```
